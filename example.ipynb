```python
import value_area as va
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("data/EURUSD_D1.csv", names=["open", "high", "low", "close", "volume"])
```

```python
df.head()
```

```python
dist = va.cal_dist(df["high"], df["low"])
vah, val, poc = va.value(dist, 70)
print(f"vah: {vah}\tval: {val}\tpoc: {poc}")
```

```python
df_dist = pd.DataFrame({"price": dist.keys(), "f": dist.values()})
df_dist[:5]
```

```python
fig, ax1 = plt.subplots(figsize=(20,12))
ax1.barh(df_dist.index, df_dist["f"]*10, color='r')
ax2 = ax1.twinx()
ax2.plot(range(len(df.index)), df["close"])
```
